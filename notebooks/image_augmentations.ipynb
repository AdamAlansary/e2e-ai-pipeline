# Image Augmentations Displayed

### Import modules

```python
import cv2
import random
import matplotlib.pyplot as plt
```

### Load image

```python
image_path = "../data/human_segmentation_kaggle/images/ds6_pexels-photo-792385.png"
mask_path = "../data/human_segmentation_kaggle/masks/ds6_pexels-photo-792385.png"

original_image = cv2.imread(image_path)
original_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)
original_width, original_height = original_image.shape[:2]

original_mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)

# Plot images
plt.figure(figsize=(14, 8))

plt.subplot(1, 2, 1)
plt.imshow(original_image)
plt.title("Original Image")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(original_mask, cmap='gray')
plt.title("Original Segmentation Mask")
plt.axis('off')

plt.show()
```

### Horizontal and vertical flipping

```python
horizontal_flip_image = cv2.flip(original_image, 1)
horizontal_flip_mask = cv2.flip(original_mask, 1)

vertical_flip_image = cv2.flip(original_image, 0)
vertical_flip_mask = cv2.flip(original_mask, 0)

# Plot images
fig, axes = plt.subplots(2, 2, figsize=(12, 12))
axes = axes.flatten()  # Now axes is a 1D array of Axes objects

axes[0].imshow(horizontal_flip_image)
axes[0].set_title("Horizontally Flipped Image")
axes[0].axis("off")

axes[1].imshow(horizontal_flip_mask, cmap="gray")
axes[1].set_title("Horizontally Flipped Segmentation Mask")
axes[1].axis("off")

axes[2].imshow(vertical_flip_image)
axes[2].set_title("Vertically Flipped Image")
axes[2].axis("off")

axes[3].imshow(vertical_flip_mask, cmap="gray")
axes[3].set_title("Vertically Flipped Segmentation Mask")
axes[3].axis("off")

plt.tight_layout()
plt.show()
```

### Gaussian blur

```python
blurred_image = cv2.GaussianBlur(original_image, (11, 11), 0)

# Plot image
plt.figure(figsize=(7, 7))

plt.imshow(blurred_image)
plt.title("Blurred Image With (11, 11) Kernel Size")
plt.axis('off')

plt.show()
```

### Downscaling (to reduce quality)

```python
# Get original image width and height
original_width, original_height = original_image.shape[:2]
# Calculate new downscaling size with factor of 0.3
new_size = (int(original_height * 0.3), int(original_width * 0.3))
# Downscale the image
downscaled = cv2.resize(original_image, new_size, interpolation=cv2.INTER_LINEAR)
# Upscale it back to the original dimensions
downscaled_image = cv2.resize(downscaled, (original_height, original_width), interpolation=cv2.INTER_LINEAR)

# Plot image
plt.figure(figsize=(7, 7))

plt.imshow(downscaled_image)
plt.title("Downscaled Image")
plt.axis('off')

plt.show()
```

### Random brightness adjustment

```python
# Factor in the range [0.5, 1.5]
factor = 0.5 + random.random()
# Apply brightness adjustment using factor
adjusted_brightness_image = cv2.convertScaleAbs(original_image, alpha=factor, beta=0)

# Plot image
plt.figure(figsize=(7, 7))

plt.imshow(adjusted_brightness_image)
plt.title(f"Adjusted Brightness Image With Factor of {factor:.3f}")
plt.axis('off')

plt.show()
```
